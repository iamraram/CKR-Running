```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle, Patch
from matplotlib.lines import Line2D

# -----------------------------------------------------------------------------------------------------

# Constants
ROWS = 4
COLS = 170  # COLS를 매우 크게 설정
MIN_DISTANCE = 10
MAX_DISTANCE = 20
MIN_DISTANCE_POTION = 15
MAX_DISTANCE_POTION = 30
MAX_HEALTH = 50
JELLY_SCORE = 18888
BEAR_JELLY_SCORE = JELLY_SCORE * 5
POTION_HEALTH = 15
OBSTACLE_PENALTY = -20
MOVE_PENALTY = -1

# -----------------------------------------------------------------------------------------------------

class CookieRunMap:
    fixed_map = None  # 클래스 변수로 고정된 맵을 정의

    def __init__(self, map_seed=None):
        self.rows = ROWS
        self.cols = COLS
        self.player_position = [0, 0]  # Player starts at the bottom row
        self.player_health = MAX_HEALTH
        self.total_score = 0
        self.jelly_count = 0
        self.bear_jelly_count = 0
        self.potion_count = 0
        self.jump_positions = []
        self.jump_count = 0
        self.health_changes = []
        self.obstacle_collisions = 0
        self.path = []  # Initialize path list to record the player's path
        self.before_health = MAX_HEALTH
        self.running = False

        if map_seed is not None:
            np.random.seed(map_seed)

        if CookieRunMap.fixed_map is None:  # 고정된 맵이 없는 경우에만 생성
            self.map = np.zeros((ROWS, COLS), dtype=int)
            self.make_map()
            CookieRunMap.fixed_map = self.map.copy()  # 생성된 맵을 클래스 변수에 저장
        else:
            self.map = CookieRunMap.fixed_map.copy()  # 고정된 맵의 복사본을 사용

    def make_map(self):
        self.map[:, :] = 0

        # Place obstacles on row 0
        col = np.random.randint(MIN_DISTANCE, MAX_DISTANCE)
        while col < self.cols:
            self.map[0, col] = 1
            self.map[1, col] = 2  # Jelly on the row above the obstacle
            col += np.random.randint(MIN_DISTANCE, MAX_DISTANCE)

        # Place jellies and bear jellies on rows 0 and 1
        for col in range(self.cols):
            if self.map[0, col] != 1:
                if np.random.rand() < 0.9:
                    self.map[0, col] = 2  # Jelly on the first row
                else:
                    self.map[0, col] = 3  # Bear Jelly on the first row
            if self.map[1, col] != 2:
                if np.random.rand() < 0.9:
                    self.map[1, col] = 2  # Jelly on the second row
                else:
                    self.map[1, col] = 3  # Bear Jelly on the second row

        # Place potions on row 1 after jellies and obstacles are placed
        col = np.random.randint(MIN_DISTANCE_POTION, MAX_DISTANCE_POTION)
        while col < self.cols:
            self.map[1, col] = 4  # Potion on the second row
            col += np.random.randint(MIN_DISTANCE_POTION, MAX_DISTANCE_POTION)

        current_row, current_col = self.player_position
        self.map[current_row, current_col] = 0

    def move_player(self, action):
        row, col = self.player_position
        self.path.append((row, col))  # Record current position to path

        if action == 'right':
            col += 1
        elif action == 'jump' and col + 2 < self.cols and row < self.rows - 1:
            # Perform a jump
            self.running = True
            self.path.append((row + 1, col + 1))  # Mid-air (before jump)
            row += 1
            col += 1
            self.update_score_and_health(row, col)  # Check and update for mid-air items
            self.path.append((row, col))  # Mid-air (after jump)
            col += 1
            row -= 1
            self.path.append((row, col))  # Landing
            self.update_score_and_health(row, col)  # Check and update for landing items
            self.jump_count += 1

            self.running = False
            self.player_health -= 1

        if col >= self.cols:
            col = self.cols - 1

        self.player_position = [row, col]
        self.apply_penalty()
        reward = self.update_score_and_health(row, col)
        self.before_health -= 1
        return reward

    def apply_penalty(self):
        row, col = self.player_position
        if self.map[row, col] == 1:
            self.before_health = self.player_health
            self.player_health += OBSTACLE_PENALTY
            self.obstacle_collisions += 1
        self.player_health += MOVE_PENALTY
        if self.player_health > MAX_HEALTH:
            self.player_health = MAX_HEALTH
        elif self.player_health < 0:
            self.player_health = 0

    def update_score_and_health(self, row=None, col=None):
        if row is None or col is None:
            row, col = self.player_position

        reward = 0

        if self.map[row, col] == 2:
            self.total_score += JELLY_SCORE
            self.jelly_count += 1
            reward += JELLY_SCORE
        elif self.map[row, col] == 3:
            self.total_score += BEAR_JELLY_SCORE
            self.bear_jelly_count += 1
            reward += BEAR_JELLY_SCORE
        elif self.map[row, col] == 4:
            self.player_health = min(MAX_HEALTH, self.player_health + POTION_HEALTH)
            self.potion_count += 1
            reward = 0
        elif self.map[row, col] == 1:
            self.player_health -= OBSTACLE_PENALTY
            reward = 0

        self.health_changes.append(self.player_health)
        self.map[row, col] = 0  # Clear the cell after collecting the item

        def check_health(a, b):
            if a > b:
                return "-" + str("{:,}".format(a - b))
            else:
                return "+" + str("{:,}".format(b - a))

        total_score_formatted = "{:,}".format(self.total_score)

        if not self.running:
            self.before_health = self.player_health

        return reward

    def is_game_over(self):
        return self.player_health <= 0

    def plot_map(self):
        fig, ax = plt.subplots(figsize=(20, 9))
        ax.set_aspect('equal')
        ax.set_xticks(np.arange(-0.5, self.cols, 1))
        ax.set_yticks(np.arange(-0.5, self.rows, 1))
        ax.grid(which='both')
        ax.set_xticklabels([])
        ax.set_yticklabels([])

        for row in range(self.rows):
            for col in range(self.cols):
                if self.map[row, col] == 1:
                    ax.add_patch(Rectangle((col - 0.5, row - 0.5), 1, 1, color='gray'))
                elif self.map[row, col] == 2:
                    ax.add_patch(Circle((col, row), 0.4, color='blue'))
                elif self.map[row, col] == 3:
                    ax.add_patch(Circle((col, row), 0.4, color='brown'))
                elif self.map[row, col] == 4:
                    ax.add_patch(Rectangle((col - 0.3, row - 0.3), 0.6, 0.6, color='orange'))

        # Draw the path if it exists
        if self.path:
            path_rows, path_cols = zip(*self.path)
            ax.plot(path_cols, path_rows, color='green', linewidth=2, alpha=0.5)

        # Draw the player position
        row, col = self.player_position
        ax.add_patch(Circle((col, row), 0.3, facecolor='green', edgecolor='white', linewidth=2))

        legend_elements = [
            Patch(facecolor='gray', edgecolor='black', label='Obstacle'),
            Circle((0, 0), 0.1, color='blue', label='Jelly'),
            Circle((0, 0), 0.1, color='brown', label='Bear Jelly'),
            Rectangle((0, 0), 0.1, 0.1, color='orange', label='Potion'),
            Line2D([0], [0], color='green', linewidth=2, label='Way')
        ]

        plt.xlim(-0.5, self.cols - 0.5)
        plt.ylim(self.rows - 0.5, -0.5)
        plt.gca().invert_yaxis()  # Ensure the bottom row is at the bottom
        plt.legend(handles=legend_elements, loc='upper right')
        plt.show()

# -----------------------------------------------------------------------------------------------------

alpha = 0.3
gamma = 0.99
epsilon = 2.0
epsilon_decay = 0.998
epsilon_min = 0.04

class QLearningAgent:
    def __init__(self, state_space, action_space, alpha=alpha, gamma=gamma, epsilon=epsilon, 콜=epsilon_decay, epsilon_min=epsilon_min):
        self.q_table = np.zeros(state_space + (len(action_space),))
        self.alpha = alpha
        self.gamma = gamma
        self.epsilon = epsilon
        self.epsilon_decay = epsilon_decay
        self.epsilon_min = epsilon_min
        self.actions = action_space

    def choose_action(self, state):
        if np.random.rand() < self.epsilon:
            return np.random.choice(self.actions)
        else:
            return self.actions[np.argmax(self.q_table[state])]

    def learn(self, state, action, reward, next_state):
        action_index = self.actions.index(action)
        predict = self.q_table[state][action_index]
        target = reward + self.gamma * np.max(self.q_table[next_state])
        self.q_table[state][action_index] += self.alpha * (target - predict)
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay

# -----------------------------------------------------------------------------------------------------

# 상태 공간과 행동 공간 정의
state_space = (ROWS, COLS, MAX_HEALTH + 1)
action_space = ['right', 'jump']

# Q-Learning 에이전트 초기화
agent = QLearningAgent(state_space, action_space)

# 학습 파라미터
num_episodes = 100
max_steps_per_episode = COLS

episode_rewards = []
ifLoaded = False

# 학습 진행
for episode in range(num_episodes):
    env = CookieRunMap()
    state = (env.player_position[0], env.player_position[1], min(env.player_health, MAX_HEALTH))
    total_reward = 0

    for step in range(max_steps_per_episode):
        action = agent.choose_action(state)
        reward = env.move_player(action)
        next_state = (env.player_position[0], env.player_position[1], min(env.player_health, MAX_HEALTH))
        agent.learn(state, action, reward, next_state)
        state = next_state
        total_reward += reward

        if env.is_game_over():
            break

    episode_rewards.append(total_reward)
    env.plot_map()
    print(f"Episode: {episode+1}, Total Reward: {'{:,}'.format(total_reward)}점, Epsilon: {agent.epsilon}", end='\n\n')


# -----------------------------------------------------------------------------------------------------

# 새로운 맵에서 추가 학습
new_map_seed = 42  # 새로운 맵 생성을 위한 시드 값
new_env = CookieRunMap(map_seed=new_map_seed)
new_episode_rewards = []

for episode in range(num_episodes):
    state = (new_env.player_position[0], new_env.player_position[1], min(new_env.player_health, MAX_HEALTH))
    total_reward = 0

    for step in range(max_steps_per_episode):
        action = agent.choose_action(state)
        reward = new_env.move_player(action)
        next_state = (new_env.player_position[0], new_env.player_position[1], min(new_env.player_health, MAX_HEALTH))
        agent.learn(state, action, reward, next_state)
        state = next_state
        total_reward += reward

        if new_env.is_game_over():
            break

    new_episode_rewards.append(total_reward)
    new_env.plot_map()
    print(f"New Map Episode: {episode+1}, Total Reward: {'{:,}'.format(total_reward)}점, Epsilon: {agent.epsilon}", end='\n\n')

# -----------------------------------------------------------------------------------------------------

# 에피소드별 점수 그래프 시각화
plt.plot(episode_rewards, label='Original Map')
plt.plot(new_episode_rewards, label='New Map')
plt.xlabel('Episode')
plt.ylabel('Total Reward')
plt.title('Total Reward per Episode')
plt.legend()
plt.show()
```
